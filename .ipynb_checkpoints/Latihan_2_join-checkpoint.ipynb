# One to many relationship

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

### 1. Report the account representative for each customer.

```python
run("""
SELECT 
    customerNumber, 
    customerName, 
    employeeNumber,
    salesRepEmployeeNumber    
FROM
    customers t1
        INNER JOIN
    employees t2 ON t1.salesRepEmployeeNumber = t2.employeeNumber
""")
```
